```python
import control as ctrl
import matplotlib.pyplot as plt

num = [1]       #numerator
den = [1, 2, 2]    # denominator s^2 + 2s + 2
G = ctrl.tf(num, den)
print(G)
```

<TransferFunction>: sys[0]

        1
  -------------
  s^2 + 2 s + 2

```python
poles = ctrl.poles(G)
print("Poles:", poles)
```

Poles: [-1.+1.j -1.-1.j]
